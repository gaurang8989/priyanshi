```python
date_list = []
for x in mycol.find({},{ "_id": 0, "actor": 1 }):
    y = x['actor']
    date_str = y["postedTime"][0:10]
#     date_time_obj = datetime.strptime(date_str, '%Y-%m-%d')
    print(date_str)
    date_list.append(date_str)
```

```python
date_list
```

```python
with open('date.csv', 'w', encoding='utf-8') as f:
        date = ','.join(date_list)
        f.write(date)
        f.close()
```

```python
from collections import Counter
dates = date_list
c = Counter()
c.update(dates) 
print(type(c))
c
```

```python
with open('date_count.txt', 'w', encoding='utf-8') as f:
        date = '\n'.join(c)
        f.write(date)
        f.close()
```
